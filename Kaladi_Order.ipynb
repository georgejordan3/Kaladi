```
import pandas as pd
import folium
# Manually import the order and zipcode data
```

```
orders = pd.read_csv("woocommerce-order-export.csv")
orders.head()
```

```
orders_zip = orders[["order_id", "billing_postcode"]]
orders_zip.head()
```

```
zip_count = orders_zip.set_index(["billing_postcode"]).value_counts("billing_postcode")
zip_count.rename(columns = {[0]: 'Count'})
zip_count

#work to count orders in each zipcode
```

In the last 1000 orders, we have received orders from 461 unique billing zipcodes.

```
denver_map = folium.Map(location=[39.7392, -104.9903], zoom_start=12)
denver_map
```

```
def map_feature_by_zipcode(zipcode_data):
    """
    Generates a folium map of Seattle
    :param zipcode_data: zipcode dataset
    :param col: feature to display
    :return: m
    """

    # read updated geo data
    denver_geo = "denverzip.geojson"

    # Initialize Folium Map with Seattle latitude and longitude
    m = folium.Map(location=[39.7392, -104.9903], zoom_start=12,
                   detect_retina=True, control_scale=False)

    # Create choropleth map
    m.choropleth(
        geo_data=denver_geo,
        name='choropleth',
        data=zip_count,
        # col: feature of interest
        #columns=['zipcode', col],
        key_on='feature.properties.ZIPCODE',
        fill_color='OrRd',
        fill_opacity=0.9,
        line_opacity=0.2,
        legend_name='Order Count'
    )

    folium.LayerControl().add_to(m)

    return m
```

```
map_feature_by_zipcode(zip_count)
#troubleshoot
```
